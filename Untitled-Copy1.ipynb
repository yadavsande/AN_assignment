```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

from sklearn.metrics import mean_squared_error, r2_score
```

### Note: There were some negative values of ordered_units in training data-set
### For the lack of further information, I assume these are returns( negative ordered units)

```python
df = pd.read_csv("base_promo_trainset (1).csv")

test_df = pd.read_csv("base_promo_testset.csv")
# Combining train and test files to to feature manipulation at one place
final_df=pd.concat([df,test_df],axis=0)
```

```python
# plot data
fig, ax = plt.subplots(figsize=(15,7))
# use unstack()
df.groupby(['date_id','MAG']).sum()['ordered_units'].unstack().plot(ax=ax)
plt.title("MAG weekly Sales(units) trendline")
plt.xlabel("Time-axis")
plt.ylabel("Ordered Units")
```

## Creating dummies for the categorical variables

```python

final_df = pd.get_dummies(final_df,columns=['product_id','MAG','AG'],drop_first=True)
final_df
```

```python
# seprating target variables and dropping account_id and date_id 
X = final_df[final_df.columns[~final_df.columns.isin(['ordered_units','account_id','date_id'])]]
y = final_df[['ordered_units']]
```

```python
print(df.shape)
print(test_df.shape)
print(final_df.shape)
```

```python
df_Train=X.iloc[:15347,:]
# this df_Test will be used later to predict the ordered_units
df_Test=X.iloc[15347:,:]
y=y.iloc[:15347,:]

```

## Spliting the data into train and test


```python
# seprating target variables and dropping account_id and date_id
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_Train,y, test_size=0.3,random_state=42)
```

## Decision Tree Regression model

```python
from sklearn.tree import DecisionTreeRegressor


regressor =DecisionTreeRegressor()
regressor.fit(X_train,y_train)

```

```python
pred_train = regressor.predict(X_train)
print('Coefficient of determination(R^2): %.2f'
      % r2_score(y_train,pred_train))
```

```python
pred = regressor.predict(X_test)
print('Coefficient of determination(R^2): %.2f'
      % r2_score(y_test,pred))
```

### Since R squred value for training data set is coming as 1, decision tree has overfitted training data,
### Overfitting can be confirmed with a very low R-squred value on test data (0.62)
### Therefore we will not use Decsion tree Regressor model

## Random Regressor Forest

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 18, random_state = 42)
regressor.fit(X_train,y_train)


pred_train = regressor.predict(X_train)
print('Coefficient of determination(R^2): %.2f'
      % r2_score(y_train,pred_train))

print('Root mean squred error(R^2): %.2f'
      % math.sqrt(mean_squared_error(y_train,pred_train)))

```

```python

pred = regressor.predict(X_test)
print('Coefficient of determination(R^2)): %.2f'
      % r2_score(y_test,pred))

print('Root mean squred error: %.2f'
      % math.sqrt(mean_squared_error(y_test,pred)))
```

## XGBoost Regressor model

```python
import xgboost as xgb
model = xgb.XGBRegressor()
model.fit(X_train,y_train)
print (model)
```

```python
pred_train = model.predict(data=X_train)
print('Coefficient of determination(R^2): %.2f'
      % r2_score(y_train,pred_train))

print('Root mean squred error: %.2f'
      % math.sqrt(mean_squared_error(y_train,pred_train)))
```

```python
pred = model.predict(data=X_test)
```

```python
print('Coefficient of determination(R^2): %.2f'
      % r2_score(y_test,pred))

print('Root mean squred error: %.2f'
      % math.sqrt(mean_squared_error(y_test,pred)))
```

## Out of the three models XGboost performs best without overfitting. 
## Let us tune hyperparameters manually to improve accuracy further
### changes are Learning rate = 0.22 , nestimator=150, max_depth=4


```python
import xgboost as xgb
model = xgb.XGBRegressor(base_score=0.5, booster='gbtree', colsample_bylevel=1,
             colsample_bynode=1, colsample_bytree=1, gamma=0,
             importance_type='gain', learning_rate=0.22, max_delta_step=0,
             max_depth=4, min_child_weight=1, missing=None, n_estimators=150,
             n_jobs=1, nthread=None, objective='reg:linear', random_state=0,
             reg_alpha=0, reg_lambda=1, scale_pos_weight=1, seed=None,
             silent=None, subsample=1, verbosity=1)
model.fit(X_train,y_train)
print (model)
```

```python
pred_train = model.predict(data=X_train)
print('Coefficient of determination(R^2): %.2f'
      % r2_score(y_train,pred_train))

print('Root mean squred error(R^2): %.2f'
      % math.sqrt(mean_squared_error(y_train,pred_train)))
```

```python
pred = model.predict(X_test)

print('Coefficient of determination(R^2): %.2f'
      % r2_score(y_test,pred))

print('Root mean squred error: %.2f'
      % math.sqrt(mean_squared_error(y_test,pred)))
```

## With manual parameter tuning both training and testing R-squred has improved. Accuracy could further be improved by 
## tuning hyperparameters using specific algorithms such as gridsearch
